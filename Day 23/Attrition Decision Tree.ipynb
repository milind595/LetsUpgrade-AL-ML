```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('dataset/general_data.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
dataset=data.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.fillna(dataset.mean())
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dataset['Attrition'] = encoder.fit_transform(dataset['Attrition'])
dataset['BusinessTravel'] = encoder.fit_transform(dataset['BusinessTravel'])
dataset['EducationField'] = encoder.fit_transform(dataset['EducationField'])
dataset['Gender'] = encoder.fit_transform(dataset['Gender'])
dataset['Department'] = encoder.fit_transform(dataset['Department'])
dataset['MaritalStatus'] = encoder.fit_transform(dataset['MaritalStatus'])
dataset['JobRole'] = encoder.fit_transform(dataset['JobRole'])

dataset.head()
```

```python
dataset1=dataset.drop(['Attrition'],axis=1)
dataset1.head()
```

```python
dataset2 = dataset.iloc[:,1:2]
dataset2.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
rf_model.fit(dataset1,dataset2)
```

```python
print("Accuracy is:",rf_model.oob_score_)
```

```python
for dataset1,imp in zip(dataset1,rf_model.feature_importances_):
    print(dataset1,imp);
```

Clearly,Age,MonthlyIncome,TotalWorkingYears are most important features and decision tree will be made based on these features only.

# Decision Tree

```python
from sklearn import tree
predictors=dataset[['Age','MonthlyIncome','TotalWorkingYears']]
treemodel=tree.DecisionTreeClassifier(max_depth=6)
treemodel.fit(X=predictors,y=dataset['Attrition'])
```

```python
with open("Attrition_Decision_Tree1.dot",'w')as f:
    f=tree.export_graphviz(treemodel,feature_names=['Age','MonthlyIncome','TotalWorkingYears'],out_file=f);
```

```python
treemodel.score(X=predictors,y=dataset["Attrition"])
```
