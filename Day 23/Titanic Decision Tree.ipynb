# Q: Build Decision Tree(DV-"Survived",IDV-"Age,Gender and Fare") and Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Dataset/train.csv")
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
dataset=data.drop(["Cabin"],axis=1)
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
dataset["Sex"]=encode.fit_transform(dataset["Sex"])
dataset.head()
```

```python
x=dataset[["Sex","Age","Fare"]]
x.head()
```

```python
y=dataset.iloc[:,1:2]
y.head()
```

```python
from sklearn import tree
predictors=x
tree_model=tree.DecisionTreeClassifier(max_depth=8)
tree_model.fit(X=predictors,y=y)
```

```python
with open('Titanic_Decision_Tree.dot','w')as f:
    f=tree.export_graphviz(tree_model,feature_names=['Sex','Age','Fare'],out_file=f);
```
