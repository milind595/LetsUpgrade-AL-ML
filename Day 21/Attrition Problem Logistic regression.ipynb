```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("general_data.csv")
data.head()
```

```python
data.info()
```

```python
data.shape
```

Dropping irrelevant attributes

Employee attrition is not effected by attributes such as as EmployeeCount and EmployeeID,moreover attributes with constant value for all the records such as Over18 and StandardHours does'nt affect the attrition rate.

```python
dataset = data.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.NumCompaniesWorked.mean()
y = dataset.TotalWorkingYears.mean()
```

```python
dataset['NumCompaniesWorked'] = dataset['NumCompaniesWorked'].fillna(x)
dataset['TotalWorkingYears'] = dataset['TotalWorkingYears'].fillna(y)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dataset['Attrition'] = encoder.fit_transform(dataset['Attrition'])
dataset['BusinessTravel'] = encoder.fit_transform(dataset['BusinessTravel'])
dataset['EducationField'] = encoder.fit_transform(dataset['EducationField'])
dataset['Gender'] = encoder.fit_transform(dataset['Gender'])
dataset['Department'] = encoder.fit_transform(dataset['Department'])
dataset['MaritalStatus'] = encoder.fit_transform(dataset['MaritalStatus'])
dataset['JobRole'] = encoder.fit_transform(dataset['JobRole'])

dataset.head()
```

```python
dataset1=dataset.drop(['Attrition'],axis=1)
dataset1.head()
```

```python
dataset2 = dataset.iloc[:,1:2]
dataset2.head()
```

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(dataset1)
logistic=sm.Logit(dataset2,x1)
result=logistic.fit()
result.summary()
```

# Inferences

1) The independent variables with p-value less than 0.05 are Age, Department, EducationField, JobRole, MaritalStatus,NumCompaniesWorked,TotalWorkingYears,TrainingTimesLastYear,YearsSinceLastPromotion,YearsWithCurrManager suggesting that all these variables play a significant role in deciding the attrition of the employee.

2) Out of all these variables, namely Age, MaritalStatus, NumCompaniesWorked, TotalWorkingYears, TrainingTimesLastYear,YearsSinceLastPromotion and YearsWithCurrManager are most significant features.

3) Out of all these variables, namely Department, EducationField, JobRole are significant features.
